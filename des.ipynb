```python
#DES KEY GENERATION

k = [int(x) for x in list("1010000111")]

def p10(k):
	permutation = [2, 4, 1, 6, 3, 9, 0, 8, 7, 5]
	temp = k.copy()
	for i in range(0, 10):
		temp[i] = k[permutation[i]]
	print("After p10 ====>  ", temp)
	return temp

def p8(k):
	permutation = [5, 2, 6, 3, 7, 4, 9, 8]
	temp = [0]*8
	for i in range(0, 8):
		temp[i] = k[permutation[i]]
	print("After p8 ====>  ", temp)
	return temp

def ls(k, start):
	temp = k[start]
	for i in range(start, start + 4):
		k[i] = k[i + 1]
	k[start + 4] = temp
	print("After Left Shift ====>  ", k)
	return

#Driver Code
o10 = p10(k)
ls(o10, 0)
ls(o10, 5)
k1 = p8(o10)
ls(o10, 0)
ls(o10, 5)
ls(o10, 0)
ls(o10, 5)
k2 = p8(o10)
print("k1 ===> ", k1)
print("k2 ===> ", k2)
```

```python
def ip(pt):
	permutation = [1, 5, 2, 0, 3, 7, 4, 6]
	temp = pt.copy()
	for i in range(0, 8):
		pt[i] = temp[permutation[i]]
	print("Initial Permutation ===> ", pt)
	return

def ep(pt):
	permutation = [3, 0, 1, 2, 1, 2, 3, 0]
	ans = [0]*8
	temp = [0]*4
	for i in range(0, 4):
		temp[i] = pt[i + 4]
	for i in range(0, 8):
		ans[i] = temp[permutation[i]]
	print("Expanded Text ===> ", ans)
	return ans

def xor(pt, k):
	for i in range(0, 8):
		pt[i] = pt[i] ^ k[i]
	print("XOR ===> ", pt)

def inverseip(pt):
	permutation = [3, 0, 2, 4, 6, 1, 7, 5]
	temp = pt.copy()
	for i in range(0, 8):
		pt[i] = temp[permutation[i]]
	print("Inverse Permutation ===> ", pt)
	return

def s0(pt, op):
	box = [
	[1, 0, 3, 2],
	[3, 2, 1, 0],
	[0, 2, 1, 3],
	[3, 1, 3, 2]]
	row = 2*pt[0] + pt[3]
	col = 2*pt[1] + pt[2]
	newsub = box[row][col]
	op[0] = (newsub//2)%2
	op[1] = newsub%2
	print("s0 ==> ", op[0], op[1])
	return

def s1(pt, op):
	box = [
	[0, 1, 2, 3],
	[2, 0, 1, 3],
	[3, 0, 1, 0],
	[2, 1, 0, 3]]
	row = 2*pt[4] + pt[7]
	col = 2*pt[5] + pt[6]
	newsub = box[row][col]
	op[2] = (newsub//2)%2
	op[3] = newsub%2
	print("s1 ==> ", op[2], op[3])
	return

def p4(op):
	permutation = [1, 3, 2, 0]
	temp = op.copy()
	for i in range(0, 4):
		op[i] = temp[permutation[i]]
	print("p4 ===> ", op)
	return

def sw(pt):
	for i in range(0, 4):
		temp = pt[i]
		pt[i] = pt[i + 4]
		pt[i + 4] = temp
	print("sw ===> ", pt)
	return

def xorlpt(pt, op):
	for i in range(0, 4):
		pt[i] = pt[i] ^ op[i]
	print("xor lpt ===> ", pt)

def fk(pt, k):
	expandedtext = ep(pt)
	xor(expandedtext, k)
	op = [0]*4
	s0(expandedtext, op)
	s1(expandedtext, op)
	p4(op)
	xorlpt(pt, op)
	return
```

```python
plaintext = [int(x) for x in list("11110111")]
originaltext = plaintext.copy()
print("Plaintext = ", plaintext)

#encryption

ip(plaintext)
print("Round 1 ")
fk(plaintext, k1)
sw(plaintext)
print("Round 2 ")
fk(plaintext, k2)
inverseip(plaintext)
print("Cipher Text ===> ", plaintext)
```

```python
#decrypt
deciphertext = plaintext.copy()
ip(deciphertext)
fk(deciphertext, k2)
sw(deciphertext)
fk(deciphertext, k1)
inverseip(deciphertext)
print("Decrypted Text ===> ", deciphertext)
for i in range(0, 8):
	if (deciphertext[i] != originaltext[i]):
		print("Failed")
		break
else:
	print("Pass")
```
